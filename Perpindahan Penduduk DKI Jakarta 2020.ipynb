## Import Libraries

```python
import pandas as pd
import os
import numpy as np
import calendar
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

## Merging all dataset

```python
all_months_data = pd.DataFrame()
files = [file for file in os.listdir('./DataPerpindahan')]
for file in files:
    df = pd.read_csv('./DataPerpindahan/'+file)
    all_months_data = pd.concat([all_months_data, df])
```

```python
all_months_data.to_csv('all_data.csv', index = False)
```

```python
all_data = pd.read_csv('all_data.csv')
all_data.head(50)
```

```python
all_data.info()
```

## Clean Up the Data

#### Find the Nan data and drop it

```python
nan_data = all_data[all_data.isna().any(axis=1)]
nan_data
```

```python
all_data = all_data.dropna()
all_data.head()
```

#### Find the unwanted value and delete it

```python
temp_df = all_data[all_data['jumlah'] == 0.0]
temp_df.head()
```

```python
all_data = all_data[all_data['jumlah']!= 0.0]
all_data.head()
```

#### Convert columns into correct data type

```python
all_data.info()
```

```python
rename_columns = ['Tahun', 'Bulan', 'Kota_Kabupaten', 'Kecamatan','Kelurahan','Jenis_Kelamin','Jumlah']
all_data.columns = rename_columns
```

```python
all_data.columns = rename_columns
```

```python
all_data.head()
```

```python
all_data['Tahun'] = pd.Categorical(all_data.Tahun)
all_data['Bulan'] = pd.Categorical(all_data.Bulan)
all_data['Kota_Kabupaten'] = pd.Categorical(all_data.Kota_Kabupaten)
all_data['Kecamatan'] = pd.Categorical(all_data.Kecamatan)
all_data['Jenis_Kelamin'] = pd.Categorical(all_data.Jenis_Kelamin)
all_data['Jumlah'] = all_data['Jumlah'].astype('int')
```

```python
all_data['Bulan'] = all_data['Bulan'].apply(lambda x: calendar.month_abbr[x])
```
